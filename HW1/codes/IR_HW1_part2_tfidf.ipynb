# Calculate IDF (documents)

```python


def compute_idf( documents , term_frequency ):
    N = len(documents)
    idf = {}
    for word , related_doc in term_frequency.items() :
      idf[word] = math.log10 (N / len(related_doc) )


    return idf
```

```python
# Compute IDF
idf = compute_idf(documents , term_frequency)
print(idf)

# Sort words by frequency
sorted_idf = sorted(idf.items(), key=lambda x: x[1], reverse=True)


top_words = [word for word, freq in sorted_idf[:9489]]
print(top_words)
```

# calculate tf-idf for documents

```python
def calculate_tfidf(term_frequency, idf, documents, terms):

    tfidf = np.zeros((len(documents), len(terms)))


    term_to_index = {term: index for index, term in enumerate(terms)}


    for word, frequency in term_frequency.items():
        if word in term_to_index:
            for doc, freq in frequency.items():
                tfidf[doc][term_to_index[word]] = freq * idf[word]

    return tfidf
```

```python
documents_list = documents['document'].tolist()

tfidf = calculate_tfidf(term_frequency, idf, documents_list, top_words)
```

# cosine similarity

```python
import numpy as np
from numpy.linalg import norm

def cosine_similarity(A, B):
    return np.dot(A, B) / (norm(A) * norm(B))
```

```python

cosine_sim = np.zeros((queries_tfidf.shape[0], tfidf.shape[0]))

# Calculate cosine similarity for each query-document pair
for i in range(queries_tfidf.shape[0]):
    for j in range(tfidf.shape[0]):
        cosine_sim[i][j] = cosine_similarity(queries_tfidf[i], tfidf[j])

```

```python
top_10_docs = {}
for i in range(cosine_sim.shape[0]):

    query_id = queries['query_id'].iloc[i]

    top_10_indices = np.argsort(cosine_sim[i])[-10:] # Get top 10 docs for each query
    top_10_doc_ids = []
    for index in top_10_indices:
        doc_id = documents['doc_id'].iloc[index]
        top_10_doc_ids.append(doc_id)
    # Store the document IDs in the dictionary using query ID
    top_10_docs[query_id] = top_10_doc_ids
```

# Evaluation

**Precision**

```python
def precision_at_k(ranked_list, positive_list, k):


    if len(ranked_list) > k:
        ranked_list = ranked_list[:k]

    precision = 0
    correct = 0

    for i, p in enumerate(ranked_list):
        if p in positive_list and p not in ranked_list[:i]:
            correct += 1
            precision += correct / (i + 1)

    if not ranked_list:
        return 0

    return precision / len(ranked_list)
```

**p@5 && p@10**

```python
p_at_5_values = []
p_at_10_values = []

for query_index, top_10_docs_indices in top_10_docs.items():
    # Get the list of positive documents
    positive_list = releation.loc[releation['query_id'] == query_index, 'doc_id'].tolist()


    # Calculate p@5 and p@10 and append the values to the respective lists
    p_at_5 = precision_at_k(top_10_docs_indices, positive_list, 5)

    p_at_10 = precision_at_k(top_10_docs_indices, positive_list, 10)
    p_at_5_values.append(p_at_5)
    p_at_10_values.append(p_at_10)

# Calculate the average p@5 and p@10 across all queries
average_p_at_5 = sum(p_at_5_values) / len(p_at_5_values)
average_p_at_10 = sum(p_at_10_values) / len(p_at_10_values)

print(f"Average p@5: {average_p_at_5}")
print(f"Average p@10: {average_p_at_10}")
```

**MAP(Mean Average Precision)**

```python
total_precision = 0
num_queries = 0

for query_index, top_10_docs_indices in top_10_docs.items():
    # Get the list of positive documents
    positive_list = releation.loc[releation['query_id'] == query_index, 'doc_id'].tolist()

    precision_sum = 0
    num_positives = len(positive_list)

    for i, doc_index in enumerate(top_10_docs_indices):
        if doc_index in positive_list:
            precision_sum += precision_at_k(top_10_docs_indices[:i+1], positive_list, i+1)

    if num_positives > 0:
        total_precision += precision_sum / num_positives
        num_queries += 1

mean_average_precision = total_precision / num_queries

print(f"Mean Average Precision (MAP): {mean_average_precision}")
```

**MRR(Mean Reciprocal Rank)**

```python
total_reciprocal_rank = 0
num_queries = 0

for query_index, top_10_docs_indices in top_10_docs.items():
    # Get the list of positive documents
    positive_list = releation.loc[releation['query_id'] == query_index, 'doc_id'].tolist()

    reciprocal_sum = 0

    for i, doc_index in enumerate(top_10_docs_indices):
        if doc_index in positive_list:
            reciprocal_sum += 1 / (i + 1)
            break

    if reciprocal_sum > 0:
        total_reciprocal_rank += reciprocal_sum
        num_queries += 1

mean_reciprocal_rank = total_reciprocal_rank / num_queries

print(f"Mean Reciprocal Rank (MRR): {mean_reciprocal_rank}")
```
