# **evaluation**

**Mean Average Precision(MAP)**

```python
def calculate_map(sorted_documents, releation):
    total_precision = 0
    num_queries = 0

    for query_id, documents in sorted_documents.items():
        # Get the list of positive documents
        positive_list = releation.loc[releation['query_id'] == query_id, 'doc_id'].tolist()

        precision_sum = 0
        num_positives = len(positive_list)

        for i, doc_id in enumerate(documents):
            if doc_id in positive_list:
                precision_sum += precision_at_k(documents[:i+1], positive_list, i+1)

        if num_positives > 0:
            total_precision += precision_sum / num_positives
            num_queries += 1

    return total_precision / num_queries


```

**precision**

```python
def precision_at_k(ranked_list, positive_list, k):

    if len(ranked_list) > k:
        ranked_list = ranked_list[:k]
  #  print(ranked_list)
    precision = 0
    correct = 0

    for i, p in enumerate(ranked_list):
        if p in positive_list and p not in ranked_list[:i]:
            correct += 1
            precision += correct / (i + 1)

    if not ranked_list:
        return 0

    return precision / len(ranked_list)
```

```python
def calculate_precision_at_k(sorted_documents, releation, k):
    total_precision = 0
    num_queries = 0

    for query_id, documents in sorted_documents.items():
        # Get the list of positive documents
        positive_list = releation.loc[releation['query_id'] == query_id, 'doc_id'].tolist()

        precision = precision_at_k(documents, positive_list, k)

        total_precision += precision
        num_queries += 1

    return total_precision / num_queries

```

**mean reciprocal rank**

```python
def calculate_mrr(sorted_documents, releation):
    total_reciprocal_rank = 0
    num_queries = 0

    for query_id, documents in sorted_documents.items():
        # Get the list of positive documents
        positive_list = releation.loc[releation['query_id'] == query_id, 'doc_id'].tolist()

        reciprocal_rank = 0

        for i, doc_id in enumerate(documents):
            if doc_id in positive_list:
                reciprocal_rank = 1 / (i + 1)
                break

        if reciprocal_rank > 0:
            total_reciprocal_rank += reciprocal_rank
            num_queries += 1

    return total_reciprocal_rank / num_queries
```

# **Retrieving documents using neural semantic vectors**

```python
import pandas as pd


documents = pd.read_csv("hw1_docs.csv")

queries = pd.read_csv("hw1_queries.csv")

releation = pd.read_csv("hw1_qrels.csv")


display(documents)
display(queries)
display(releation)
```

```python
import pandas as pd
import numpy as np
import re
from gensim.utils import simple_preprocess

from gensim.parsing.preprocessing import STOPWORDS


def preprocess(text):
 # Convert text to lowercase
 text = text.lower()

 # Remove special characters
 text = re.sub(r'\W', ' ', text)

 # Remove extra spaces
 text = re.sub(r'\s+', ' ', text)

 # Tokenize the text
 words = text.split()

 # Remove stop words
 words = [word for word in words if word not in STOPWORDS]

 # Join the words back into a string
 preprocessed_text = ' '.join(words)

 return preprocessed_text


```

```python
queries['preprocessed_text'] = queries['query'].apply(preprocess)
documents['preprocessed_text'] = documents['document'].apply(preprocess)
```

# bert model

```python
from transformers import AutoTokenizer, AutoModel
import torch
from tqdm import tqdm

# Load pre-trained model tokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased' )

# Load pre-trained model
model = AutoModel.from_pretrained('bert-base-uncased' , output_hidden_states=True)

# Check if CUDA is available and set PyTorch device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Move the model to the GPU
model = model.to(device)



```

```python


# Create semantic vectors for queries
query_vectors = {}
for index, row in tqdm(queries.iterrows(), total=queries.shape[0]):
  inputs = tokenizer(row['preprocessed_text'], return_tensors='pt', truncation=True, padding='max_length', max_length=512)
  inputs = inputs.to(device) # Move inputs to GPU
  outputs = model(**inputs)
  query_vectors[row['query_id']] = outputs.last_hidden_state.mean(dim=1).detach().cpu().numpy() # Average the hidden states of all tokens


# Create semantic vectors for documents
doc_vectors = {}
for index, row in tqdm(documents.iterrows(), total=documents.shape[0]):
  inputs = tokenizer(row['preprocessed_text'], return_tensors='pt', truncation=True, padding='max_length', max_length=512)
  inputs = inputs.to(device) # Move inputs to GPU
  outputs = model(**inputs)
  doc_vectors[row['doc_id']] = outputs.last_hidden_state.mean(dim=1).detach().cpu().numpy() # Average the hidden states of all tokens
```

```python
print(query_vectors)
```

```python
import numpy as np
def cosine_similarity(a, b):
  a = np.squeeze(a)
  b = np.squeeze(b)
  dot_product = np.dot(a, b)
  norm_a = np.linalg.norm(a)
  norm_b = np.linalg.norm(b)
  return dot_product / (norm_a * norm_b)
```

```python
# Calculate cosine similarity between query vector and each document vector
similarities = {}
sorted_docs = {}
for query_id, query_vector in query_vectors.items():
   similarities[query_id] = {}
   for doc_id, doc_vector in doc_vectors.items():
       similarity = cosine_similarity(query_vector, doc_vector)
       similarities[query_id][doc_id] = similarity

# Sort documents based on their similarity to each query
top_10_docs = {}
for query_id, doc_similarities in similarities.items():
   sorted_docs[query_id] = sorted(doc_similarities.items(), key=lambda x: x[1], reverse=True)


sorted_docs_bert = {}
for query_id, sorted_docs_query in sorted_docs.items():
  sorted_docs_bert[query_id] = [doc_id for doc_id, similarity in sorted_docs_query[:10]]
```
